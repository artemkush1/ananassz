```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
def f(x,u, const):
    return -const[0]**2*x-const[1]*u+const[2]*x**3
```

```python
print('input const')
const=list(map(float, input().split(',')))
print('input baseline', 'x_0=', sep='\n', end='')
x = [float(input())]
print('u_0=', end='')
u = [float(input())]
h=np.linspace(0.0, 5*2*pi/const[0], 10**(6))
w=h[1]-h[0]
for t_i in h[1:]:
    u.append(u[-1] + w * f(x[-1], u[-1], const))
    x.append(x[-1] + w * u[-2])
plt.plot(h, x, '-')
```
