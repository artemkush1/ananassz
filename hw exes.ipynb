A snail curve is defined by the following equation in polar coordinates:  ùëü=ùëé+cosùúô. Plot the curve on an xy plane for several values of  ùëé. (Hint: np.linspace creates a linearly spaced array of values)

```python
%matplotlib notebook
from math import sin, cos 
np.linspace(-10,10,20)
phu=np.linspace(0,10,100)
ap=np.linspace(0,1,100)
#plt.plot(1, 1, "o", color = "red" )
#phi=b[1,1]
z=0
for a in range(1, 4):
    for phi in phu:
        z+=1
        r=a/2+cos(phi)
        if z%2:
            plt.plot(r*cos(phi), r*sin(phi), "o", color = "lavender" )
        else:
            plt.plot(r*cos(phi), r*sin(phi), "o", color = "pink" )
```
