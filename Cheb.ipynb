```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

```python
def Lg(x, k, X_k):
    y = 1
    for i in range (len(X_k)) :
        if i != k:
            y*=((x - X_k[i])/ (X_k[k]-X_k[i]))
    return y

def F(X_new, X_s, Y_s):
    return sum (Y_s[i]*Lg(X_new, i, X_s) for i in range(len(X_s)))
```

```python
def Cheb_knots(knots):
    x_data = [(np.cos(np.pi*(2*n - 1)/(2*knots))) for n in range(1, knots+1)]
    y_data = [1/(25*i**2 + 1) for i in x_data]
    x_list = np.linspace (-1, 1, 500)
    f=[(F(i, x_data, y_data)) for i in x_list]
    plt.plot (x_data, y_data, "o", x_list, f, '--')
    plt.show()
```

```python
Cheb_knots(15)
```
